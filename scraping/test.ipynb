```python
import random
import requests
import re

url = 'https://ja.wikipedia.org/wiki/%E3%82%A2%E3%83%A1%E3%83%AA%E3%82%AB%E5%90%88%E8%A1%86%E5%9B%BD%E3%81%A8%E6%97%A5%E6%9C%AC%E3%81%AE%E9%96%93%E3%81%AE%E5%A7%89%E5%A6%B9%E9%83%BD%E5%B8%82%E6%8F%90%E6%90%BA%E3%81%AE%E4%B8%80%E8%A6%A7'

res = requests.get(url)
res.encoding = res.apparent_encoding  # 文字コードを設定

from bs4 import BeautifulSoup

soup = BeautifulSoup(res.text, "lxml")
title = soup.title.text 
```

```python
print(title)
```

```python
text = []
data = []
for table1 in soup.find_all('table', attrs={"class":"wikitable"}):
    for tr in table1.find_all('tr'):
        for td in tr.find_all('td'):
            row = td.text.split('\n')
            text.append(td.text.split('\n'))
            #print(row)
            
regex = re.compile(r'(\d{4})年*')

i = 0
while i < len(text):
    if i+2 < len(text):
        years = regex.match(text[i+2][0])
        if years != None:
            year = years.group(1)
        else:
            year = "?"
        data.append([text[i][0], text[i+1][0], year])
    i += 3
data

```
